## Load libraries:

```python
%matplotlib inline
import numpy as np
import asciidata as asd
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('small')
```

## Load the data file and organize its columns into vectors:

```python
#file='/disks/shear7/piladiez/INT-ui/catalogues/fits/isocontours_simplebroken_chi2red_0.2DMbin.cat'
file='/disks/shear7/piladiez/INT-ui/catalogues/fits/isocontours_complexbroken_chi2red_0.2DMbin.cat'
cat=asd.open(file)
```

```python
Q_IN=cat[0].tonumpy()
Q_OUT=cat[1].tonumpy()
N_IN=cat[2].tonumpy()
N_OUT=cat[3].tonumpy()
RHO_IN=cat[4].tonumpy()
ERR_RHO_IN=cat[5].tonumpy()
RHO_OUT=cat[6].tonumpy()
ERR_RHO_OUT=cat[7].tonumpy()
CHI2_IN=cat[8].tonumpy()
CHI2_OUT=cat[9].tonumpy()
```

## Mapping the minimum $chi^2_{red}$ across the pair of grid parameters for each of the two axisymmetric laws (inner and outer):

```python
from lib_frange import frange
nin_step,nout_step,q_step=0.1,0.2,0.05
RANGE_nin=frange(-1,-4.1,nin_step)
RANGE_nout=frange(-3,-7.1,nout_step)
RANGE_q=frange(0.3,1.6,q_step)
print len(RANGE_nin),len(RANGE_nout),len(RANGE_q)
```

* Power law index ($n$) and polar axis ratio ($q$) for the inner region:

```python
ARRAY_chi2red_nq_in=np.zeros((len(RANGE_q),len(RANGE_nin)))
ARRAY_rho_nq_in=np.zeros((len(RANGE_q),len(RANGE_nin)))
ARRAY_Erho_nq_in=np.zeros((len(RANGE_q),len(RANGE_nin)))
LIST_q_in, LIST_n_in=[],[]

Nn=0
for n_in in reversed(RANGE_nin):
    use_n = np.where(N_IN==round(n_in,2))
    SUBSET_q= Q_IN[use_n]
    SUBSET_n= N_IN[use_n]
    SUBSET_chi2=CHI2_IN[use_n]
    SUBSET_rho=RHO_IN[use_n]
    SUBSET_err_rho=ERR_RHO_IN[use_n]
    
    Nq=0
    for q_in in reversed(RANGE_q):
        use_q = np.where(SUBSET_q==round(q_in,2))
        SUBSET2_q= SUBSET_q[use_q]
        SUBSET2_n= SUBSET_n[use_q]
        SUBSET2_chi2=SUBSET_chi2[use_q]
        SUBSET2_rho=SUBSET_rho[use_q]
        SUBSET2_err_rho=SUBSET_err_rho[use_q]
        #print n,q,len(SUBSET2_chi2)
        ARRAY_chi2red_nq_in[Nq][Nn]=np.min(SUBSET2_chi2)
        ARRAY_rho_nq_in[Nq][Nn]=SUBSET2_rho[np.argmin(SUBSET2_chi2)]
        ARRAY_Erho_nq_in[Nq][Nn]=SUBSET2_err_rho[np.argmin(SUBSET2_chi2)]
        LIST_q_in.append(q_in)
        Nq=Nq+1
    LIST_n_in.append(n_in)
    Nn=Nn+1

#print np.min(ARRAY_chi2red_nn),np.max(ARRAY_chi2red_nn)
print 'Min =',np.min(ARRAY_chi2red_nq_in)#,np.max(ARRAY_chi2red_ninq)
i,j = np.unravel_index(ARRAY_chi2red_nq_in.argmin(), ARRAY_chi2red_nq_in.shape)
print 'at q=',LIST_q_in[i],', n=',LIST_n_in[j]
print 'for rho0=',ARRAY_rho_nq_in[i][j]
#print np.shape(ARRAY_chi2red_nn)
```

```python
import matplotlib
fontP.set_size('xx-small')

from matplotlib.colors import LogNorm

plt.figure(num=1, figsize=(10,10))
plt.tick_params(labelsize=16) 
#plt.title('$\chi^2_{red}$ map for $n$ and $q$')


plt.xlabel('power law index, $n$', fontsize=14)
plt.ylabel('polar axis ratio, $q$', fontsize=14)

cmap = matplotlib.cm.spectral
im=plt.imshow(ARRAY_chi2red_nq_in,extent=[min(RANGE_nin),max(RANGE_nin)+nin_step,min(RANGE_q),max(RANGE_q)+q_step], 
              aspect='auto',cmap=cmap)#, interpolation='nearest')#, norm=LogNorm()
              #,aspect='auto')

CS1 = plt.contour(ARRAY_chi2red_nq_in, levels=frange(np.min(ARRAY_chi2red_nq_in)+0.1, np.min(ARRAY_chi2red_nq_in)+0.7, 0.1), 
                 colors='black', #origin='lower',linewidths=2,
                 extent=[min(RANGE_nin),max(RANGE_nin)+nin_step,max(RANGE_q)+q_step,min(RANGE_q)])
plt.clabel(CS1, inline=1,fmt='%1.2f', fontsize=12)


CS3 = plt.contour(ARRAY_chi2red_nq_in, levels=frange(np.min(ARRAY_chi2red_nq_in)+1.1, np.max(ARRAY_chi2red_nq_in), 0.5), 
                 colors='black', #origin='lower',linewidths=2,
                 extent=[min(RANGE_nin),max(RANGE_nin)+nin_step,max(RANGE_q)+q_step,min(RANGE_q)])
plt.clabel(CS3, inline=1,fmt='%1.2f', fontsize=12)

# We can still add a colorbar for the image, too.
CBI = plt.colorbar(im, orientation='horizontal', shrink=0.8, label='$\chi^2_{red}$')
#CBI = plt.colorbar(im, orientation='horizontal', shrink=0.8)

plt.plot(LIST_n_in[j],LIST_q_in[i],'r+',markersize=14)
plt.plot(LIST_n_in[j]+nin_step/2,LIST_q_in[i]+q_step/2,'r.',markersize=10)


#ax = CBI.ax
#text = ax.yaxis.label
#font=matplotlib.font_manager.FontProperties(size=16)
#text.set_font_properties(font)

plt.show()
```

* Power law index ($n$) and polar axis ratio ($q$) for the outer region:

```python
ARRAY_chi2red_nq_out=np.zeros((len(RANGE_q),len(RANGE_nout)))
ARRAY_rho_nq_out=np.zeros((len(RANGE_q),len(RANGE_nout)))
ARRAY_Erho_nq_out=np.zeros((len(RANGE_q),len(RANGE_nout)))
LIST_q_out, LIST_n_out=[],[]

Nn=0
for n_out in reversed(RANGE_nout):
    use_n = np.where(N_OUT==round(n_out,2))
    SUBSET_q= Q_OUT[use_n]
    SUBSET_n= N_OUT[use_n]
    SUBSET_chi2=CHI2_OUT[use_n]
    SUBSET_rho=RHO_OUT[use_n]
    SUBSET_err_rho=ERR_RHO_OUT[use_n]
    
    Nq=0
    for q_out in reversed(RANGE_q):
        use_q = np.where(SUBSET_q==round(q_out,2))
        SUBSET2_q= SUBSET_q[use_q]
        SUBSET2_n= SUBSET_n[use_q]
        SUBSET2_chi2=SUBSET_chi2[use_q]
        SUBSET2_rho=SUBSET_rho[use_q]
        SUBSET2_err_rho=SUBSET_err_rho[use_q]
        #print n,q,len(SUBSET2_chi2)
        ARRAY_chi2red_nq_out[Nq][Nn]=np.min(SUBSET2_chi2)
        ARRAY_rho_nq_out[Nq][Nn]=SUBSET2_rho[np.argmin(SUBSET2_chi2)]
        ARRAY_Erho_nq_out[Nq][Nn]=SUBSET2_err_rho[np.argmin(SUBSET2_chi2)]
        LIST_q_out.append(q_out)
        Nq=Nq+1
    LIST_n_out.append(n_out)
    Nn=Nn+1

#print np.min(ARRAY_chi2red_nn),np.max(ARRAY_chi2red_nn)
print 'Min =',np.min(ARRAY_chi2red_nq_out)#,np.max(ARRAY_chi2red_ninq)
k,l = np.unravel_index(ARRAY_chi2red_nq_out.argmin(), ARRAY_chi2red_nq_out.shape)
print 'at q=',LIST_q_out[k],', n=',LIST_n_out[l]
print 'for rho0=',ARRAY_rho_nq_out[k][l]
#print np.shape(ARRAY_chi2red_nn)
```

```python
import matplotlib
fontP.set_size('xx-small')

from matplotlib.colors import LogNorm

plt.figure(num=1, figsize=(10,10))
plt.tick_params(labelsize=16) 
#plt.title('$\chi^2_{red}$ map for $n$ and $q$')


plt.xlabel('power law index, $n$', fontsize=14)
plt.ylabel('polar axis ratio, $q$', fontsize=14)

cmap = matplotlib.cm.spectral
im=plt.imshow(ARRAY_chi2red_nq_out,extent=[min(RANGE_nout),max(RANGE_nout)+nout_step,min(RANGE_q),max(RANGE_q)+q_step], 
              aspect='auto',cmap=cmap)#, interpolation='nearest')#, norm=LogNorm()
              #,aspect='auto')

CS1 = plt.contour(ARRAY_chi2red_nq_out, levels=frange(np.min(ARRAY_chi2red_nq_out)+0.1, np.min(ARRAY_chi2red_nq_out)+0.7, 0.1), 
                 colors='black', #origin='lower',linewidths=2,
                 extent=[min(RANGE_nout),max(RANGE_nout)+nout_step,max(RANGE_q)+q_step,min(RANGE_q)])
plt.clabel(CS1, inline=1,fmt='%1.2f', fontsize=12)


CS3 = plt.contour(ARRAY_chi2red_nq_out, levels=frange(np.min(ARRAY_chi2red_nq_out)+1.1, np.max(ARRAY_chi2red_nq_out), 0.5), 
                 colors='black', #origin='lower',linewidths=2,
                 extent=[min(RANGE_nout),max(RANGE_nout)+nout_step,max(RANGE_q)+q_step,min(RANGE_q)])
plt.clabel(CS3, inline=1,fmt='%1.2f', fontsize=12)

# We can still add a colorbar for the image, too.
CBI = plt.colorbar(im, orientation='horizontal', shrink=0.8, label='$\chi^2_{red}$')
#CBI = plt.colorbar(im, orientation='horizontal', shrink=0.8)

plt.plot(LIST_n_out[l],LIST_q_out[k],'r+',markersize=14)
plt.plot(LIST_n_out[l]+nout_step/2,LIST_q_out[k]+q_step/2,'r.',markersize=10)


#ax = CBI.ax
#text = ax.yaxis.label
#font=matplotlib.font_manager.FontProperties(size=16)
#text.set_font_properties(font)

plt.show()
```

## Comparing the best fit parameters resulting from the minimum in the grid to the those resulting from a continues range of values:

```python
qlocus_chi2min_in = LIST_q_in[i]
nlocus_chi2min_in = LIST_n_in[j]

qlocus_chi2min_out = LIST_q_out[k]
nlocus_chi2min_out = LIST_n_out[l]

print 'INNER AXISYMMETRIC LAW:'
print 'Locus for the min(chi^2_red):'
print 'q=',qlocus_chi2min_in,'pm',q_step/2,'  n=',nlocus_chi2min_in,'pm',nin_step/2

print 'with rho0=','%s' % float('%.2g' % ARRAY_rho_nq_in[i][j]),'pm','%s' % float('%.1g' % ARRAY_Erho_nq_in[i][j])
print 'and chisquare_red=','%s' % float('%.4g' % np.min(ARRAY_chi2red_nq_in))


print ''
print 'OUTER AXISYMMETRIC LAW:'
print 'Locus for the min(chi^2_red):'
print 'q=',qlocus_chi2min_out,'pm',q_step/2,'  n=',nlocus_chi2min_out,'pm',nout_step/2

print 'with rho0=','%s' % float('%.2g' % ARRAY_rho_nq_out[k][l]),'pm','%s' % float('%.1g' % ARRAY_Erho_nq_out[k][l])
print 'and chisquare_red=','%s' % float('%.4g' % np.min(ARRAY_chi2red_nq_out))
```

This compares to the best fit returned by the non-grid script as follows:

The best fit found by the script returns:

* INNER: 

$q=0.50 \pm 0.36$,  $n=-3.5 \pm 1.0$

with $\rho_0=0.001 \pm 0.003$

and $\chi^2_{red}=0.820$


* OUTER:

$q=0.75 \pm 0.10$,  $n=-5.3 \pm 0.2$

with $\rho_0=0.18 \pm 0.15$

and $\chi^2_{red}=1.481$


*The uncertainties in the case of the inner law are so large that we have to conclude the parameters are practically unconstrained. This is probably due to the small amount of data available for the inner region (22 data points in 8 or less different lines of sight). The reasonable value of $\chi^2_{red}$ indicates that the curve fits nicely enough the data (as we can see in the fitted profiles plots), but it does not reflect the problem with the best fit parameters. Nontheless and ignoring the large uncertainties, the best fit values match reasonably well the best results from the grid if we take into account the grid steps as cell uncertainties. This suggests that the uncertainties in the best fits could probably be reduced.*


*The parameters in the outer case seem to be reasonably constrained by the uncertainties. Once we compare them to the results from the grid we find good agreement within the uncertainties. $\rho_0$ should probably be dismissed since it is not possible to constrain the inner kiloparsec of the halo with outer data. The difference in the $\chi^2_{red}$ is largely due to the difference in the number of free parameters. Indeed if we calculate the one of the best fit based on 1 free parameter ($rho_0$), we would get $1.42$, which is smaller but still not comparable to that of the grid.*
